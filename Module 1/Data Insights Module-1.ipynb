```python
#Lets Visualize the Online and Offline orders
Trans_Online_Order = Transactions['online_order']
ax = sns.countplot(Trans_Online_Order)

total = len(Trans_Online_Order)
for x in ax.patches:
    height = x.get_height()
    ax.text(x.get_x()+x.get_width()/2,
            height,
            '{:.2f}%'.format(height/total*100),
            ha = 'center')
    
# 1 FOR TRUE AND  0 FOR FALSE
```
